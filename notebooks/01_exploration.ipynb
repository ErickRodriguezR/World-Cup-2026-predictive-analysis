```python
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()

df_matches = pd.read_csv("../data/raw/WorldCupMatches.csv")
df_cups = pd.read_csv("../data/raw/WorldCups.csv")
```

```python
print("Matches:", df_matches.head())
print("Cups:", df_cups.head())

df_matches.head()
df_cups.head()

API_KEY = os.getenv("FOOTBALL_API_KEY")
```

```python
total_matches = df_matches.shape[0] #all the matches in all World Cup
missing_values = df_matches.isnull().sum() #empty values in the dataset
year_range = df_matches["Year"].max() - df_matches["Year"].min() #range of 1930 to 2014

home_teams = df_matches["Home Team Name"]
away_teams = df_matches["Away Team Name"]
unique_teams = pd.concat([home_teams, away_teams]).nunique() #sum of all the unique teams

df_matches["total_goals"] = df_matches["Home Team Goals"] + df_matches["Away Team Goals"]
average_goals = df_matches["total_goals"].mean() #average goals per match

print(total_matches)
print(missing_values)
print(year_range)
print(unique_teams)
print(average_goals)
```

```python
import os
import requests
from dotenv import load_dotenv

load_dotenv("../.env")

api_key = os.getenv("FOOTBALL_API_KEY")
headers = {
    "X-Auth-Token": api_key
}
url = "https://api.football-data.org/v4/competitions/WC/matches"

response = requests.get(url, headers=headers)
data = response.json()

api_matches = pd.DataFrame(data.get("matches", []))

api_matches.head()
```
