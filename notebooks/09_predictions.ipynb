```python
import pandas as pd
import numpy as np
import joblib

# Load model
model = joblib.load("../outputs/final_model.pkl")
```

```python
# Load 2026 World Cup teams data (already scaled by notebook 07)
df_2026 = pd.read_csv("../data/raw/rankings_2026.csv")

print(f"Loaded {len(df_2026)} teams")
df_2026.head()
```

```python
# Features (excluding 'team' column)
FEATURES = [
    "matches_played", "goals_for", "goals_against", "wins", "draws", "losses",
    "average_attendance", "goals_per_match", "win_percentage", "goal_difference",
    "confederation_AFC", "confederation_CAF", "confederation_CONCACAF",
    "confederation_CONMEBOL", "confederation_OFC", "confederation_Other", "confederation_UEFA"
]

X_2026 = df_2026[FEATURES]

# Generate predictions
probabilities = model.predict_proba(X_2026)[:, 1]

# Create results table
results = pd.DataFrame({
    "team": df_2026["team"],
    "probability": (probabilities * 100).round(1)
}).sort_values("probability", ascending=False)

results["ranking"] = range(1, len(results) + 1)

print("=" * 70)
print("WORLD CUP 2026 - PREDICTED FINALISTS (TOP 2)")
print("=" * 70)

finalists = results.head(2)
print(f"\n{finalists.iloc[0]['team']} - {finalists.iloc[0]['probability']}% probability")
print(f"{finalists.iloc[1]['team']} - {finalists.iloc[1]['probability']}% probability")

print("\n" + "=" * 70)
print("HEAD-TO-HEAD FINAL")
print("=" * 70)
print(f"\n   {finalists.iloc[0]['team']}  vs  {finalists.iloc[1]['team']}")
```

```python
# Show top 10 teams
print("=" * 70)
print("TOP 10 TEAMS - RANKED BY PROBABILITY")
print("=" * 70)
print(results.head(10).to_string(index=False))

# Show all 48 teams
print("\n" + "=" * 70)
print("ALL 48 TEAMS - RANKED BY PROBABILITY")
print("=" * 70)
print(results.to_string(index=False))
```

```python
# Save predictions to CSV
output_path = "../data/processed/predictions_2026.csv"
results.to_csv(output_path, index=False)

print(f"Predictions saved to: {output_path}")
print(f"Total teams: {len(results)}")
print(f"Predicted finalists: {finalists.iloc[0]['team']}, {finalists.iloc[1]['team']}")
```
